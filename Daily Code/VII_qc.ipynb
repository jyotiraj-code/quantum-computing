```python
#Writing the function called query

def query(elements = [1], marked_elements = [0]):
    for i in marked_elements:
        elements[i] = -1 * elements[i]

    return elements
```

```python
#Writing a function called inversion
def inversion(elements = [1]):
    summation = 0
    
    for i in range(len(elements)):
        summation += elements[i]

    #The mean value
        mean = summation / len(elements)

    #Reflection over mean
        for i in range(len(elements)):
            value = elements[i]
            new_value = mean - (value - mean)
            elements[i] = new_value

        return elements
```

```python
#Now we will execute the program for N = 16
#And when the first element is marked for 5 steps

#Importing dependencies
from math import asin, pi

#Initial values

iteration = 5
N = 16
marked_elements = [0]
k = len(marked_elements)
elements = []
state_on_unit_circle = []

#Initial quantum state
for i in range(N):
    elements.append(1/N**0.5)

x = elements[N-1] * ((N-k)**0.5)
y = elements[0] *(k**0.5)
state_on_unit_circle.append([x,y,"0"])


#Executing the Grover's search algorithm for iteration steps

for step in range(iteration):
    #query
    elements = query(elements, marked_elements)
    x = elements[N-1] * ((N-k)**0.5)
    y = elements[0] * (k**0.5)
    state_on_unit_circle.append([x,y,str(step) + "''"])

    #inversion
    elements = inversion(elements)
    x = elements[N-1] * ((N-k)**0.5)
    y = elements[0] * (k**0.5)
    state_on_unit_circle.append([x,y,str(step+1)])


    #Printing the angles
    print("angles in degree")
    for state in state_on_unit_circle:
        print(asin(state[1])/pi*180)



```
